# <font color=Red> Import libraries</font>

```python
import pandas as pd # For files
import matplotlib.pyplot as plt # For graph
```

# <font color=Red> Read the CSV file</font>

```python
file_path = 'Fire alarm system\Data\smoke_detection_iot.csv'
data = pd.read_csv(file_path)
```

# <font color='Red'>Analyzing data to "fire" and "non-fire"</font>

```python
# Based on the last column in file "the target", I will be divided it into two types, "fire" and "non_fire"
fire = data[data['Fire_Alarm'] == 1]
non_fire = data[data['Fire_Alarm'] == 0]
```

# <font color=Red> Calculate the count of fire alarms </font>

```python
# I calculate the number of the fire alarm and non-fire alarm to draw the graph to be simple to understand.
fire_alarms_count = len(fire)
non_fire_alarms_count = len(non_fire)
```

# <font color=Red> Create a pie chart</font>

```python

sizes = [fire_alarms_count, non_fire_alarms_count]
my_labels = ['Yes_Fire', 'No_Fire']

#size: parameter is defined array of my data. ("fire_alarms_count","non_fire_alarms_count")
#labels: parameter is defined names of my data.
#autopct: parameter used to indicate the numerical ratio of each wedge.
#counterclock: Specify fractions direction, clockwise or counterclockwise. hint: I make it false to be like photo in task
#stratangle: parameter is defined with an angle in degrees.
#shadow: make the graph like 3d.

plt.pie(sizes, labels = my_labels, autopct= '%1.1f%%', counterclock= False, startangle = 90, shadow = True)

#To add a list of explanation for each wedge
plt.legend()

# Add a title for the graph
plt.title('Fire Alarm Graph')

# Display the pie chart
plt.show()
```
